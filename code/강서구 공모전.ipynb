```python
TotalData['1인당 복지수준'] = TotalData['시설합계'] / TotalData['65세이상 인구_y']

TotalData['독거노인 비율'] = TotalData['독거노인 총인구수']/TotalData['65세이상 인구_y']

TotalData['65세이상 인구 비율'] = TotalData['65세이상 인구_y']/TotalData['전체인구_y']

TotalData['1인당 요양수준'] = TotalData['노인 요양 시설수'] / TotalData['65세이상 인구_y']

TotalData['기초수급노인 비율'] = TotalData['기초수급노인']  / TotalData['65세이상 인구_y']

TotalData['복지시설 수준'] = TotalData['종사자수']  / TotalData['시설합계']


TotalData['요양시설 수준'] = TotalData['요양시설 종사자수']  / TotalData['노인 요양 시설수']


TotalData = TotalData.applymap(lambda x: float(x))
TotalData.head()
```
